# Anomaly Detection Analysis with LOF

## Overview
This document outlines the process of analyzing network traffic data using the Local Outlier Factor (LOF) method. The analysis involves reading multiple CSV files containing network flow data, processing it to generate communication paths between IP addresses, and visualizing the results with pie charts. 

## Dependencies
The analysis requires the `pandas` library for data manipulation and analysis. Ensure that you have it installed in your Python environment.

## Logic and Approach

**Data Loading:** The code first loads network flow data from CSV files into pandas DataFrames. This is essential to access and manipulate the data for further analysis.

**Creating Communication Paths:** By concatenating source_ip and destination_ip, the code creates a new representation of conversations. This step is crucial for analyzing the flow of communication and understanding how different IPs interact.

**Counting and Visualization:** The use of value_counts() helps in identifying the most common communication paths, and the pie charts provide a visual representation of these counts, making it easier to interpret the data.

**Repetition for Multiple Datasets:** The same processing steps are repeated for three different datasets, ensuring consistency in analysis and allowing for comparisons across different scenarios or datasets.

This analysis provides valuable insights into the network traffic patterns based on the communication between IP addresses. The visualization of conversation flows through pie charts allows for an easier understanding of the data distribution and can help identify potential anomalies in network behavior. By following this methodology, practitioners can better monitor and analyze network traffic, ultimately aiding in anomaly detection and cybersecurity efforts.


```python
import pandas as pd
```

```python
lof_001_allfeat = pd.read_csv("test_result_lof_001_allfeat.csv", encoding="utf8",low_memory = False)
lof_001_ipca = pd.read_csv("test_result_lof_001_ipca.csv", encoding="utf8",low_memory = False)
lof_001_var = pd.read_csv("test_result_lof_001_var.csv", encoding="utf8",low_memory = False)
```

```python
lof_001_allfeat['conversations'] = lof_001_allfeat[['source_ip','destination_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_allfeat.conversations.value_counts()[:15]
```

```python
lof_001_allfeat.conversations.value_counts()[:15].plot.pie()
```

```python
lof_001_ipca['conversations'] = lof_001_ipca[['source_ip','destination_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_ipca.destination_ip.value_counts()
#.head(60)[:15]
```

```python
lof_001_ipca.conversations.value_counts()[:15].plot.pie()
```

```python
lof_001_var['conversations'] = lof_001_var[['source_ip','destination_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_var.conversations.value_counts()[:15]
```

```python
lof_001_var.conversations.value_counts()[:15].plot.pie()
```
